# A Demo for MrFiRA

Server starts on request. Contact: qjt163@163.com

```python
import requests

target_url = "https://mrfira.cpolar.cn/run"

def retrieve(code, query, topk):
    payload = {
        "code": f"{code}",
        "query": f"{query}",
        "topk": f"{topk}"
    }

    try:
        response = requests.post(target_url, json=payload)
        return response.text
    except Exception as e:
        return e
```

### Example 1: The company's total assets at the end of 2022

```python
code = "300126"                     # company code
query = "该公司2022年末的资产总额"
topk = 20                           # send top k reranked evidence to LLM

response = retrieve(code, query, topk)
print(response)
```
